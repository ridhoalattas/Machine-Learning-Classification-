```
import pandas as pd
import os
```

```
os.listdir('.')
# os.chdir('Dataset')
```

```
dataset = pd.read_csv('iris.csv')
```

```
feature = dataset.drop(['Id', 'Species'], axis = 1)
label = dataset['Species']
```

```
from sklearn import preprocessing
from sklearn.metrics import f1_score, precision_score, recall_score
scaler = preprocessing.StandardScaler()

feature = scaler.fit_transform(feature.values)
label = scaler.fit_transform(label.values.reshape(-1,1)).flatten()
```

**No. 1**

```
from sklearn.model_selection import train_test_split
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size=0.15)
```

```
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators = 50, bootstrap=True, max_features="sqrt")
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)
print("Nilai Estimator : 50\n")

print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

```
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators = 100, bootstrap=True, max_features="sqrt")
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)
print("Nilai Estimator : 100\n")

print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

```
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators = 200, bootstrap=True, max_features="sqrt")
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)
print("Nilai Estimator : 200\n")

print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

Dari kesimpulan pengujian 3 nilai estimator diatas, didapatkan **nilai estimator 50, 100 dan 200 hasil akurasinya sama dengan nilai estimator 300**

**No. 2**

```
from sklearn.model_selection import train_test_split
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size=0.15, random_state = 10)
```

```
from sklearn.svm import SVC

svm_classifier = SVC(kernel = 'poly')
svm_classifier.fit(feature_train, label_train)

predictions = svm_classifier.predict(feature_train)

print("Polynomial Kernel : \n")
print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

```
from sklearn.svm import SVC

svm_classifier = SVC(kernel = 'linear')
svm_classifier.fit(feature_train, label_train)

predictions = svm_classifier.predict(feature_train)

print("Linear Kernel : \n")
print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

```
from sklearn.svm import LinearSVC

lsvc = LinearSVC()
lsvc.fit(feature_train, label_train)

predictions = lsvc.predict(feature_train)

print("LinearSVC Kernel : \n")
print(f1_score(label_train, predictions, average='macro'))
print(precision_score(label_train, predictions, average='macro'))
print(recall_score(label_train, predictions, average='macro'))
```

Dari pengujian diatas, didapatkan bahwa akurasi yang didapat dengan menggunakan **SVC linear dan SVC polynomial adalah sama**.
Sedangkan akurasi yang didapatkan dengan **LinearSVC lebih kecil sedikit dibandingkan dengan polynomial dan linear**.

**No. 3**

```
from sklearn.model_selection import train_test_split
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size=0.15)
```

```
from sklearn.neighbors import KNeighborsClassifier

for i in range(1, 16, 2) :
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(feature_train, label_train)
  predictions = knn.predict(feature_train)
  
  print(f"\nn_neighbors : {i}")
  print(f1_score(label_train, predictions, average='macro'))
  print(precision_score(label_train, predictions, average='macro'))
  print(recall_score(label_train, predictions, average='macro'))
```

**Nilai n_neighbors dengan model k-NN dengan akurasi terbaik adalah 1.**
